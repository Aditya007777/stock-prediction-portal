```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Type of Plots

## Line Plots

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]
```

```python
plt.plot(x,y);
plt.title('Simple Line plot');
plt.xlabel('X axis');
plt.ylabel('Y axis');
plt.grid(True)
```

## Bar Chart

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
revenue = [100, 200, 300, 500]
```

```python

plt.figure(figsize=(10, 6))
plt.bar(months, revenue)
plt.title('Monthly Revenue Bar Chart');
plt.xlabel('Months');
plt.ylabel('Revenue');
```

## Scatter plots

```python
heights = [65, 70, 68, 72, 66]
weights = [150, 160, 155, 180, 145]
```

## Subplots And Legends 

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.legend()
plt.title('Plot with legends')
plt.xlabel('Weekdays')
plt.ylabel('Price')
```

```python
#First subplot
plt.subplot(3,1,1)
plt.plot(days,original_price,'b-o', label='original price')
plt.legend()
plt.title('Plot with days vs original price')
plt.xlabel('Weekdays')
plt.ylabel('Price')


#Second subplot
plt.subplot(3,1,2)
plt.plot(days,predicted_price,'r-o', label='predicted price')
plt.legend()
plt.title('Plot with days vs predicted price')
plt.xlabel('Weekdays')
plt.ylabel('Price')

#Third subplot
plt.subplot(3,1,3)
difference_price = [original_price[i] - predicted_price[i]  for i in range(len(original_price))]
plt.bar(days,difference_price, color = 'purple', label='price differnce')
plt.title('Plot with days vs original price')
plt.xlabel('Weekdays')
plt.ylabel('Price')

plt.tight_layout()
```

### Saving the figure

## Saving the figure

```python
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.legend()
plt.title('Plot with legends')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.savefig('example_prediction.png', dpi=300) # save the image
```
